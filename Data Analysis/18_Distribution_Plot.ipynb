```python
import seaborn as sns
import numpy as np

np.random.seed(0)
x=np.random.randn(100)
ax_1=sns.distplot(x)
```

```python
ax_2=sns.distplot(x,rug=True,hist=False)
```

```python
ax_3=sns.distplot(x,rug=True,hist=True)
```

```python
from scipy.stats import norm
ax_4=sns.distplot(x,fit=norm)
```

```python
ax_5=sns.distplot(x,fit=norm, kde=False)
```

```python
# vertical
ax_6=sns.distplot(x, vertical=True)
```

```python
# COLOR
sns.set_color_codes()
ax_7=sns.distplot(x, color='r')
```

```python
# BACKGROUND
sns.set(style='dark')
ax_8=sns.distplot(x, color='g')
```
