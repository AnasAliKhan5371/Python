```python
import pandas as pd
path="C:/Users/Anas Ali Khan/OneDrive/Desktop/cars.xlsx"
df=pd.read_excel(path)
print(df)
```

```python
headers=["s.no","even","odd","multiple 5","nultiple 3","multiple 4","Length"]
```

```python
df.columns=headers
df.head(5)
```

```python
df["multiple 4"]/=2
df.rename({"multiple 4":"multiple 2"})
```

```python
df.rename(columns={"multiple 4":"multiple 2"},inplace=True)
df.head(5)
```

```python
df["multiple 2"]=df["multiple 2"].astype("int")
df.head(5)
```

```python
# simple feature scaling method
df["Length"]=df["Length"]/df["Length"].max()
df.head(5)
```

```python
# min-max method
df["Length"]=(df["Length"]-df["Length"].min())/(df["Length"].max()-df["Length"].min())
df.head(5)
```

```python
# z-score method
df["Length"]=(df["Length"]-df["Length"].mean())/(df["Length"].std())
df.head(5)
```
