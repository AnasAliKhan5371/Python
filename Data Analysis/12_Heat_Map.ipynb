```python
import numpy as np
np.random.seed
import seaborn as sns
uniform_data=np.random.rand(5,6)
print(uniform_data)
```

```python
heat_map=sns.heatmap(uniform_data)
```

```python
heat_map_1=sns.heatmap(uniform_data, vmin=0,vmax=1)
```

```python
normal_data=np.random.rand(6,6)
print(normal_data)
```

```python
heat_map_3=sns.heatmap(normal_data,center=0)
```

```python
flights=sns.load_dataset("flights")
flights=flights.pivot(index="month",columns="year",values="passengers")
heat_map_4=sns.heatmap(flights)
```

```python
heat_map_5=sns.heatmap(flights,annot=True,fmt="d")
```

```python
# color scheme
heat_map_6=sns.heatmap(flights,cmap='YlGnBu')
```

```python
heat_map_7=sns.heatmap(flights,cmap='RdBu')
```

```python
# remove bar
heat_map_8=sns.heatmap(flights,cbar=False)
```
